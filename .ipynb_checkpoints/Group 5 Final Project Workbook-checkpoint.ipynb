## Predicting Airline Customer Satisfaction using Machine Learning

##### Let's import our data and the libraries needed to work with our data

```python
# warnings reported for function updates, ignore them
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('airline_passenger_satisfaction.csv')
```

```python
data.head()
```

##### Key points about our data:
* Our target variable is satisfaction, which has 2 labels ("satisfied" or "neutral or dissatisfied")
* We have 4 categorical features, and the rest are numerical

### Step 1) Data cleaning

```python
# our first column contains an index for each row, which is not required
data_cleaned = data.drop(columns=['Unnamed: 0'])
```

```python
# checking for missing values
data_cleaned.isnull().sum()
```

##### The "arrival_delay_in_minutes" column is missing data in 393 rows. This requires imputation. Let's see which metric should be used to replace the missing values, mean or median.

```python
data_cleaned['arrival_delay_in_minutes'].mean()
```

```python
data_cleaned['arrival_delay_in_minutes'].median()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data_cleaned['arrival_delay_in_minutes'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Histogram of Arrival Delay in Minutes')
plt.xlabel('Arrival Delay (Minutes)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

##### The median of the arrival delay column is 0, and the mean is 15.09, and in this case it is more appropriate to impute the missing values with the mean. Let's do that now.

```python
mean_arrival_delay = data_cleaned['arrival_delay_in_minutes'].mean()
data_cleaned['arrival_delay_in_minutes'] = data_cleaned['arrival_delay_in_minutes'].fillna(mean_arrival_delay)
```

##### Let's check the data types of our columns and make sure they are all correct.

```python
data_cleaned.dtypes
```

##### The arrival delay column is a float, and the departure delay is an int. Let's change arrival delay to int as well.

```python
data_cleaned['arrival_delay_in_minutes'] = data_cleaned['arrival_delay_in_minutes'].astype(int)
```

##### Let's encode our categorical features now so they can be fed into our model properly.

```python
categorical_features = ['Gender', 'customer_type', 'type_of_travel', 'customer_class']
data_encoded = pd.get_dummies(data_cleaned, columns=categorical_features, drop_first=True)
```

##### Let's also encode our target feature.

```python
data_encoded['satisfaction'] = data_encoded['satisfaction'].map({
    'satisfied': 1,
    'neutral or dissatisfied': 0
})
```

```python
data_encoded.head()
```

##### It seems our customer loyalty column has defaulted to showing us True or False for disloyal customers, whereas we want our values to show us True or False for loyal customers. Let's change that.

```python
data_encoded['customer_type_loyal Customer'] = ~data_encoded['customer_type_disloyal Customer']
data_encoded['customer_type_loyal Customer'] = data_encoded['customer_type_loyal Customer'].astype(bool)
```

```python
data_encoded.head()
```

```python
comparison_df = data_encoded[['customer_type_disloyal Customer', 'customer_type_loyal Customer']].copy()
comparison_df['are_opposites'] = comparison_df['customer_type_disloyal Customer'] == ~comparison_df['customer_type_loyal Customer']
comparison_df.head()
```

##### Our changes have been made correctly, so let's drop our "customer_type_disloyal Customer" column

```python
data_encoded.drop(columns=['customer_type_disloyal Customer'], inplace=True)
```

##### Our target feature is relatively balanced.

### Step 3) Feature Scaling

##### To determine ranges for each feature in the dataset

```python
# Select numerical columns
numerical_columns = data_encoded.select_dtypes(include=['float64', 'int64']).columns

# Get the minimum and maximum values for each numerical column
ranges = data[numerical_columns].agg(['min', 'max']).T
ranges.columns = ['Min', 'Max']

# Display the ranges
print(ranges)
```

##### Peforming feature scaling on the numerical features to normalize their values for modelling

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Step 1: Selecting columns for scaling
columns_to_scale = ['flight_distance', 'departure_delay_in_minutes', 'arrival_delay_in_minutes']

# Step 2: MinMax Scaling
minmax_scaler = MinMaxScaler()
data_minmax_scaled = minmax_scaler.fit_transform(data_encoded[columns_to_scale])

# Converting back to a DataFrame for readability
data_minmax_scaled_df = pd.DataFrame(data_minmax_scaled, columns=columns_to_scale)

# Step 3: Standard Scaling
standard_scaler = StandardScaler()
data_standard_scaled = standard_scaler.fit_transform(data_encoded[columns_to_scale])

# Converting back to a DataFrame for readability
data_standard_scaled_df = pd.DataFrame(data_standard_scaled, columns=columns_to_scale)

# Display the results
print("MinMax Scaled Data:")
print(data_minmax_scaled_df)
print("\nStandard Scaled Data:")
print(data_standard_scaled_df)
```

### Step 4) Modeling

##### Now that we have done feature scaling, we can move on to modeling to determine the best model for predicting airline satisfaction. We will be using cross-validation to train and validate the model to prevent overfitting.

Let's set up the train and test splits and scale them for modeling

```python
from sklearn.model_selection import train_test_split

X = data_encoded.drop(columns=['satisfaction']) #feature column
y = data_encoded['satisfaction'] #target column

# Split the data into training and test sets (80% training and 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Apply scaling to training and test data
scaled_X_train = minmax_scaler.fit_transform(X_train)
scaled_X_test = minmax_scaler.fit_transform(X_test)
```

#### Baseline Models
The Logistic Regression and Decision Tree models will give us a baseline perfomance to compare against the more advanced models.

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

# Initialize and train
logreg = LogisticRegression(max_iter=10000, random_state=42)
logreg.fit(scaled_X_train, y_train)

# Cross-validation
logreg_cv = cross_val_score(logreg, scaled_X_train, y_train, cv=5, scoring='accuracy')
print(f"Logistic Regression Mean CV Score: {logreg_cv.mean():.4f}")
print(f"Logistic Regression CV Score - Standard Deviation: {logreg_cv.std():.4f}")
```

##### The Logistic Regression model has a mean CV (cross-validation) score of 0.8747 and a standard deviation across 5 folds of 0.0022.

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize and train
dtree = DecisionTreeClassifier(max_depth=10, min_samples_split=20, max_leaf_nodes=15, random_state=42)
dtree.fit(scaled_X_train, y_train)

# Cross-validation
dtree_cv = cross_val_score(dtree, scaled_X_train, y_train, cv=5, scoring='accuracy')
print(f"Decision Tree Mean CV Score: {dtree_cv.mean():.4f}")
print(f"Decision Tree CV Score - Standard Deviation: {dtree_cv.std():.4f}")
```

##### The Decision Tree model has a mean CV score of 0.9090 and a standard deviation across 5 folds of 0.0020.

##### So far, the Decicion Tree model has a better performance than the Logistic Regression model based on the mean CV score and variance (standard deviation).

##### Advanced Models
Now we'll train 3 advanced models and also evalute them using cross-validation.

#### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train
rf = RandomForestClassifier(n_estimators=50, max_depth=10, min_samples_split=20, max_leaf_nodes=20, random_state=42)
rf.fit(scaled_X_train, y_train)

# Cross-validation
rf_cv = cross_val_score(rf, scaled_X_train, y_train, cv=5, scoring='accuracy')
print(f"Random Forest Classifier Mean CV Score: {rf_cv.mean():.4f}")
print(f"Random Forest Classifier CV Score - Standard Deviation: {rf_cv.std():.4f}")
```

##### The Random Forest Classifier model has a mean CV score of 0.9207 and a standard deviation across 5 folds of 0.0018.

#### XGBoost Classifier

```python
import xgboost as xgb

# Initialize and train
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
xgb_model.fit(scaled_X_train, y_train)

# Cross-validation
xgb_model_cv = cross_val_score(xgb_model, scaled_X_train, y_train, cv=5, scoring='accuracy')
print(f"XGBoost Classifier Mean CV Score: {xgb_model_cv.mean():.4f}")
print(f"XGBoost Classifier CV Score - Standard Deviation: {xgb_model_cv.std():.4f}")
```

##### The XGBoost Classifier model has a mean CV score of 0.9620 and a standard deviation across 5 folds of 0.0024.

#### Neural Network (MLPClassifier)

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score

# Initialize and train with reduced max_iter, parallelized cross-validation, and a simpler model
nn = MLPClassifier(random_state=42, max_iter=100, tol=1e-4, hidden_layer_sizes=(50,), solver='sgd')

# Cross-validation with parallelization
nn_cv = cross_val_score(nn, scaled_X_train, y_train, cv=3, scoring='accuracy', n_jobs=-1)

# Output results
print(f"Neural Network (MLPClassifier) Mean CV Score: {nn_cv.mean():.4f}")
print(f"Neural Network (MLPClassifier) CV Score - Standard Deviation: {nn_cv.std():.4f}")
```

##### Out of all the trained models, the XGBoost Classifier model has the best performance with the highest mean CV score (0.9620) even with its high variance (0.0024).

### Evaluation:

__Logistic Regression Evaluation:__

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score

print('Logistic Regression Predictions: ', logreg.predict(scaled_X_test))
print('Logistic Regression score for training set:',logreg.score(scaled_X_train,y_train,sample_weight=None))
print('Logistic Regression score for testing set:',logreg.score(scaled_X_test,y_test,sample_weight=None))
print('Intercept: ', logreg.intercept_, '\nCoefficients: ',logreg.coef_[0])


y_pred = logreg.predict(scaled_X_test)
y_proba=logreg.predict_proba(scaled_X_test)[:, 1]
accuracy=accuracy_score(y_pred,y_test)
print('\nAccuracy score:', accuracy) #Accuracy score: 56%
print()

# Classification report
print(classification_report(y_test, y_pred))

# Confusion matrix
print(confusion_matrix(y_test, y_pred))

# ROC-AUC score
roc_auc = roc_auc_score(y_test, y_proba)
print("ROC-AUC Score:", roc_auc)
```
